```python
#he creado una funcion para poder ver las filas nulas de los elementos
# principalmente para ver patrones de porque hay nulos y como crear una
#decision para crear valores

def visualizar_nulos(attacks, columna, head=50):

    valores_nulos = attacks[columna].isnull()
    
    filas_con_nulos = attacks[valores_nulos]
    
    print(filas_con_nulos.head(head))


```

```python
visualizar_nulos(data_sin_nulos, 'Area', head=100)
```

```python
visualizar_nulos(data_sin_nulos, 'Location', head=100)
```

```python
visualizar_nulos(data_sin_nulos, 'Activity', head=100)
```
