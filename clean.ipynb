```python
#he creado una funcion para poder ver las filas nulas de los elementos
# principalmente para ver patrones de porque hay nulos y como crear una
#decision para crear valores

def visualizar_nulos(dataframe, columna):
    # identificar filas con valores nulos en la columna
    valores_nulos = dataframe[columna].isnull()
    
    # Filtrar valores nulos de columnas
    filas_con_nulos = dataframe[valores_nulos]
    
    # Mostrar las primeras 50 filas
    print(filas_con_nulos.head(50))




```

```python
# aqui el nombre no es relevante pero crearemos una funcion para que
## introduzca female en caso de ser mujer y male en caso de ser hombre

def subs_null_names(attacks):
    # sustituir null con male si sex es hombre
    attacks.loc[(attacks['Sex'] == 'M') & attacks['Name'].isnull(), 'Name'] = 'male'
    
    # sustituir null con female si sex es mujer
    attacks.loc[(attacks['Sex'] == 'F') & attacks['Name'].isnull(), 'Name'] = 'female'
```

```python
def unknown_injury(attacks):
    # Obtener una lista de los únicas
    unique_activities = attacks['Activity'].unique()

    # Crear un diccionario para relacionar actividades con tipos de lesiones
    activity_to_injury = {Activity: f"{Activity.lower()} injury" for Activity in unique_activities}

    # concatenar el valor de la columna 'Injury' con el valor de la columna 'Activity' seguido de "injury" si 'Injury' no es nulo
    attacks['Injury'] =attacks['Injury'].combine_first(attacks['Activity'] + ' injury')

    # Reemplazar el valor nulo en 'Injury' con "unknown injury" si la actividad no está en el diccionario
    attacks.loc[attacks['Injury'].isnull(), 'Injury'] = attacks['Activity'].map(activity_to_injury).fillna('unknown injury')

unknown_injury(data_sin_nulos)
```

```python
## pasamos a diccionario para contar las veces que se repiten los diferentes tipos de filas
species_dict = data_sin_nulos['Species'].value_counts().to_dict()

species_dict
```

```python
## elimino los numeros  simbolos, corchetes por nada

clean_species_dict = {re.sub(r'[0-9\'.\[\]]', '', key).strip(): value for key, value in species_dict.items()}

clean_species_dict
```

```python
clean_species_dict = {}

for key, value in species_dict.items():
    # eliminar todo los que no preceda a shark.
    if re.search(r'\b\w+\sshark\b', key):
        # eliminar numeros y corchetes
        clean_key = re.sub(r'[0-9\[\]\']', '', key).strip()
        # eliminar a y m que este entre spacios
        clean_key = re.sub(r'\ba\s|\bm\s', '', clean_key).strip()
        # eliminar espacios y puntos
        clean_key = re.sub(r'\.\s', '', clean_key)
        # eliminar la palabra to
        clean_key = re.sub(r'\bto\b', '', clean_key).strip()
        #eliminar todo lo que este a la izquierda de la coma
        clean_key = clean_key.split(',')[0].strip()
        clean_species_dict[clean_key] = value

clean_species_dict
```

```python
for index, row in data_sin_nulos.iterrows():
    species_name = row['Species']
    # insertar los valores del diccionario a la tabla de nuevo
    if species_name in clean_species_dict:
        data_sin_nulos.at[index, 'Species'] = clean_species_dict[species_name]
```

```python
z_threshold = 3

# Select only the numeric columns for outlier detection
numeric_columns = data_sin_nulos.select_dtypes(include=[np.number])

# Calculate the z-scores for each numeric column
z_scores = np.abs((numeric_columns - numeric_columns.mean()) / numeric_columns.std())

# Identify outliers by checking if the z-scores exceed the threshold
outliers_mask = z_scores > z_threshold

# Filter the DataFrame to show only the rows with outliers
outliers_data = data_sin_nulos[outliers_mask.any(axis=1)]

# Display the rows containing outliers
outliers_data
```

```python
# Calculate z-scores for numeric columns
z_threshold = 3
numeric_columns = data_sin_nulos.select_dtypes(include=np.number).columns
z_scores = np.abs((data_sin_nulos[numeric_columns] - data_sin_nulos[numeric_columns].mean()) / data_sin_nulos[numeric_columns].std())

# Create a Boolean mask to identify rows with outliers
outliers_mask = (z_scores > z_threshold).any(axis=1)

# Filter the DataFrame to keep only rows without outliers
data_sin_nulos_no_outliers = data_sin_nulos[~outliers_mask]

# Reset the index of the DataFrame if desired
data_sin_nulos_no_outliers.reset_index(drop=True, inplace=True)

# Display the DataFrame without outliers
data_sin_nulos_no_outliers
```
